```python
#LOADING THE DATASET
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#importing the dataset file from google colab
#from google.colab import files
#uploaded = files.upload()

file_name = '~' #Ensure that your csv file is in same directory as your script or else put the full path in the file name.

#reading the dataset file into a pandas data frame
data = pd.read_csv(file_name)

#displaying the data frame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

data
#NaN values are the missing values

#CHECKING THE DATAFRAME FOR MISSING VALUES
#Columns to check for missing values
columns_to_check = ['Actual_Temperature', 'Predicted_Temperature_Day1', 'Predicted_Temperature_Day2', 'Predicted_Temperature_Day3']

#Check for null values across the specified columns
null_data = data[columns_to_check].isnull()

#Displaying the rows with null values in the specified columns
null_data_rows = data[null_data.any(axis=1)]
print(null_data_rows)

#Number of missing values across all three columns
missing_values = data.isnull().sum()
print(missing_values)

#MISSING VALUE IMPUTATION USING LINEAR INTERPOLATION
#Using linear interpolation to find the missing values, it uses the neighbouring data points to calculate the missing value
data_interpolated = data.interpolate(method='linear')

# Save the cleaned DataFrame to a new CSV file
data_interpolated.to_csv('Name_of_Cleaned_File.csv', index=False)

# Displaying the cleaned DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
data_interpolated

#EXPLORATORY DATA ANALYSIS - (MEAN, MEDIAN, MODE ANALYSIS)
#uploaded = files.upload() 
#uncomment the above code if you are using google colab

file_name = 'Name_of_Cleaned_File.csv'
df = pd.read_csv(file_name)

#Calculate mean, median, and mode of the actual temperatures
mean_actual_temp = df['Actual_Temperature'].mean()
median_actual_temp = df['Actual_Temperature'].median()
mode_actual_temp = df['Actual_Temperature'].mode()

#Display the calculated values
print(f"Mean of Actual Temperatures: {mean_actual_temp}")
print(f"Median of Actual Temperatures: {median_actual_temp}")
print(f"Mode of Actual Temperatures: {mode_actual_temp}")

#Create a histogram of the actual temperatures
plt.figure(figsize=(10, 6))
plt.hist(df['Actual_Temperature'].dropna(), bins=20, edgecolor='black')
plt.title('Histogram of Actual Temperatures')
plt.xlabel('Actual Temperature')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
